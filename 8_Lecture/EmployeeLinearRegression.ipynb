```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_csv('EmployeeData.csv')
print(data)
```

```python
x = data['WorkingHours'].values.reshape(-1, 1)
y = data['incentive']
plt.scatter(x,y,marker="*",color='indigo')
plt.xlabel("Working Hours")
plt.ylabel("Incentive")
plt.title("Working Hours vs Incentive")
plt.grid()
```

```python
x_train,x_test,y_train,y_test = train_test_split(
    x,y,test_size=0.2,random_state=42
)
print("\nTraining data size", len(x_train))
print("Testing data size", len(x_test))
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
plt.scatter(x, y, color='red', label="Actual Line")
plt.plot(x,model.predict(x),color='indigo',label="Regression Line")
plt.xlabel("Working Hours")
plt.ylabel("Incentive")
plt.title("Linear Regression Fit")
plt.legend()
plt.grid()
```

```python
predicted_marks = model.predict([[18]])

print("\n Predicted Incentive for 12 Working hours:", predicted_marks[0])
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Load data from csv and print last 5 records of it
df = pd.read_csv("pokemon_data.csv")
print(df.tail(5))
```

```python
# 2. Find the empty records and update it with ‘FILLED’ value in Type2 field.
df['Type2'].fillna('FILLED', inplace=True)
print(df)
```

```python
# 3. Calculate median on Defense field.
median_defense = df['Defense'].median()
print("Median Defense:", median_defense)
```

```python
# 4. Find average of HP, Attack and Defense field and display it.
average_values = df[['HP', 'Attack', 'Defense']].mean()
print(f'Average of HP, Attack, and Defense:\n{average_values}')
```

```python
# 5. Display bar chart with name and attack fields.
plt.figure(figsize=(10,6))
plt.bar(df['Name'], df['Attack'], color="indigo")
plt.title("name and attack chart")
plt.xlabel("Name")
plt.ylabel("Attack")
plt.xticks(rotation=90)
plt.grid()
```

```python
# 6. Find sum of speed using numpy.
speed_sum = np.sum(df['Speed'])
print("Sum of Speed:",speed_sum)
```

```python
# 7. Display Legendary field where value is “True”.
value_True = df[df['Legendary'] == True]
print(value_True)
```

```python
# 8. Display Histogram on Generation.
plt.figure(figsize=(8, 6))
plt.hist(df['Generation'], edgecolor='black')
plt.title('Histogram: Generation Distribution')
plt.xlabel('Generation')
plt.ylabel('Frequency')
plt.show()
```

```python
# 9. Find Standard Deviation of speed field
std_speed = df['Speed'].std()
print(std_speed)
```

```python
# 10. Draw a scatter plot HP and Speed fields
sns.scatterplot(x='HP', y='Speed', data=df)
plt.title('Scatter Plot: HP vs Speed')
plt.xlabel('HP')
plt.ylabel('Speed')
plt.grid()
```

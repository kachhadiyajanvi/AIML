```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Question 1
df = pd.read_csv("customer.csv")
print(df.head(10))
```

```python
# Question 2
df_sorted_by_age = df.sort_values(by='Age', ascending=True)
print(df_sorted_by_age)
```

```python
# Question 3
df_cleaned = df.dropna()
print(df)
```

```python
# Question 4
age = df['Age'].values
balance = df['Balance'].values
plt.plot(age, balance, marker='o', linestyle='-', color='b')
plt.title('Line Chart: Age vs Balance')
plt.xlabel('Age')
plt.ylabel('Balance')
plt.grid()
```

```python
# Question 5
df_sorted_by_CreditScore = df.sort_values(by='CreditScore', ascending=False)
print(df_sorted_by_CreditScore.head(15))
```

```python
# Question 6
geography_counts = df['Geography'].value_counts()
custom_colors = ['#ff9999', '#66b3ff', '#99ff99']
plt.figure(figsize=(7, 7))
plt.pie(geography_counts, labels=geography_counts.index, autopct='%1.1f%%', startangle=140, colors=custom_colors)
plt.title('Pie Chart: Geography Distribution')
plt.legend()
plt.show()
```

```python
# Question 7
high_salary_customers = df[df['EstimatedSalary'] > 75000]
print(high_salary_customers)
```

```python
# Question 8
df_mean = df['Balance'].mean()
print("Mean balance:",df_mean)
```

```python
# Question 9
credit_score_permutations = np.random.permutation(df['CreditScore'])
print(credit_score_permutations[:10])
```

```python
# Question 10
sns.scatterplot(x='Geography', y='Balance', data=df, hue='Geography')
plt.title('Scatter Plot: Geography vs Balance')
plt.xlabel('Geography')
plt.ylabel('Balance')
plt.grid()
```

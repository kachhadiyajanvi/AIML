```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Load the weather dataset and show column names.
df = pd.read_csv("weatherHistory.csv")
print(df.columns)
```

```python
# 2. Display last 5 records.
print(df.tail())
```

```python
# 3. Convert date column to datetime format.
df['DateColumnName'] = pd.to_datetime(df['Formatted Date'])
print(df.head())
```

```python
# 4. Find average temperature.
avg_temp = df["Temperature (C)"].mean()
print("avg_temp:", avg_temp)
```

```python
# 5. Find days with temperature above average.
above_average_days = df[df['Temperature (C)'] > avg_temp]
print(above_average_days[['Temperature (C)','DateColumnName']].head(10))
```

```python
# 6. Calculate maximum and minimum humidity.
max_humidity = df['Humidity'].max()
print(max_humidity)

min_humidity = df['Humidity'].min()
print(min_humidity)
```

```python
# 7. Count records by weather condition.
count = df["Summary"].value_counts()
print(count)
```

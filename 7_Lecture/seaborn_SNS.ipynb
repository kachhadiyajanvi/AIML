```python
flights = sns.load_dataset("flights")
print(flights)
flights_pivot = flights.pivot(
    index='month',
    columns='year',
    values = 'passengers'
)
print(flights_pivot)

#plot heatmap
plt.figure(figsize=(12,6))
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d")
plt.title("Monthly Airline Passengers Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```

```python
car_crashes = sns.load_dataset("car_crashes")
numeric_cols = car_crashes.select_dtypes(include=["number"])
corr_matrix = numeric_cols.corr()
print(corr_matrix)

# #plot heatmap
plt.figure(figsize=(12,6))
sns.heatmap(corr_matrix, cmap="YlGnBu", annot=True, fmt=".2f")
plt.title("Corrlation Heatmap of car Crashes Dataset")
plt.show()
```

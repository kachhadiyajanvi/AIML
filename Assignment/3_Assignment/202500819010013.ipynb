```python
# Section C Assignment
# Question 1
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [12000, 15000, 18000, 17000, 21000, 25000]

plt.plot(months, sales, marker='*', linestyle='-', color='b')
plt.show()
```

```python
# Question 2
plt.plot(months, sales, marker='*', linestyle='-', color='b')
plt.title('Monthly Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.grid()
```

```python
# Question 3
plt.plot(months, sales, marker='s', linestyle=':', color='b')
plt.title('Monthly Sales Data')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.grid()
```

```python
# Question 4
plt.bar(months, sales, color='green')
plt.title('Monthly Sales Data (Bar Chart)')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.show()
```

```python
# Question 5
plt.pie(sales, labels=months, autopct='%1.1f%%', startangle=90)
plt.title('Sales Percentage by Month')
plt.show()
```

```python
# Question 6
plt.bar(months, sales, color='brown')
plt.title('Monthly Sales Data (Bar Chart)')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.grid()
```

```python
# Question 7
plt.scatter(months, sales, color='r', marker='x')
plt.title('Scatter Plot of Sales Data')
plt.xlabel('Month Index')
plt.ylabel('Sales ($)')
plt.show()
```

```python
# Question 8
plt.hist(sales, bins=5, color='purple', edgecolor='black', alpha=0.5)
plt.title('Sales Histogram')
plt.xlabel('Sales ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Question 9
plt.plot(months, sales, marker='o', label='Sales (Line)', color='b')
plt.plot(months, [x * 0.8 for x in sales], marker='x', label='80% Sales', color='r', linestyle='--')
plt.title('Multiple Lines in One Chart')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.legend()
plt.grid()
```

```python
# Question 10
plt.figure(figsize=(10, 6))
plt.plot(months, sales, marker='o', color='g', label='Sales')
plt.title('Sales Line Chart')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.grid()
plt.savefig('sales_chart.jpg')
```

```python
# Section D Assignment
# Question 1
df = pd.read_csv('categoryData.csv')
print(df)
```

```python
# Question 2
grouped_data = df.groupby('Category')['Sales'].sum()
print(grouped_data)
```

```python
# Question 3
grouped_data.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales ($)')
plt.grid()
```

```python
# Question 4
df['Sales'].plot(kind='hist', bins=10, color='green', alpha=0.7, edgecolor='black')
plt.title('Sales Distribution')
plt.xlabel('Sales ($)')
plt.ylabel('Frequency')
plt.grid()
```

```python
# Question 5
df['Sales'].plot(kind='box', color='red', vert=False)
plt.title('Boxplot of Sales')
plt.xlabel('Sales ($)')
plt.grid()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Que 1:
df = pd.read_csv("ML Datasets/ecommerce.csv")
print("First 5 records:")
print(df.head())
```

```python
#Que 2:
df = df.dropna(subset=['Customer ID'])
```

```python
#Que 3:
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
#Que 4:
df['TotalAmount'] = df['Quantity'] * df['Price']
```

```python
# Que 5:
total_sales_country = df.groupby('Country')['TotalAmount'].sum().sort_values(ascending=False)
print("Total sales by country:")
print(total_sales_country)
```

```python
# Que 6:
top_customers = df.groupby('Customer ID')['TotalAmount'].sum().sort_values(ascending=False).head(10)
print("\nTop 10 customers by total purchase value:")
print(top_customers)
```

```python
# Que 7:
negative_quantity = df[df['Quantity'] < 0]
print("Invoices with negative quantity:")
print(negative_quantity)
```

```python
# Que 8:
average_total = df['TotalAmount'].mean()
print("Average TotalAmount:", average_total)
```

```python
# Que 9:
above_average = df[df['TotalAmount'] > average_total]
print("Transactions with TotalAmount above average:")
print(above_average)
```

```python
# Que 10:
plt.figure(figsize=(10,6))
sns.boxplot(x=df['TotalAmount'])
plt.title('Box plot of TotalAmount')
plt.xlabel('TotalAmount')
plt.show()
```

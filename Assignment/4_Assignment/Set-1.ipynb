```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("ML Datasets/Churn_Modelling.csv")
#Que 1:
print(df.shape)
print(df.columns.tolist())
```

```python
#Que 2:
print(df.sample(8))
```

```python
#Que 3:
balance_avg = df['Balance'].mean()
print(balance_avg)
above_data = df[df['Balance'] > balance_avg]
print(above_data)
```

```python
#Que 4:
customers_countBy_geography = df['Geography'].value_counts()
print(customers_countBy_geography)
```

```python
#Que 5:
mean_credit = df['CreditScore'].mean()
median_credit = df['CreditScore'].median()
variance_credit = df['CreditScore'].var()
print(f"Mean: {mean_credit}, Median: {median_credit}, Variance: {variance_credit}")
```

```python
#Que 6:
sns.scatterplot(data=df, x='Age', y='Balance', hue='Gender')
plt.title('Age vs Balance by Gender')
plt.grid()
```

```python
#Que 7:
X = df[['Age']]  # Predictor
y = df['Balance']  # Target

model = LinearRegression()
model.fit(X, y)

print("\nLinear Regression model:")
print(f"Intercept: {model.intercept_}, Coefficient: {model.coef_[0]}")
```

```python
# Que 8:
age_40 = np.array([[40]])
predicted_balance = model.predict(age_40)
print(f"\nPredicted Balance for Age 40: {predicted_balance[0]}")
```

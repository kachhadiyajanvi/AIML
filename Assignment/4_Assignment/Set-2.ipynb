```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ML Datasets/StudentsPerformance.csv")
#Que 1:
print(df.columns.tolist())
```

```python
#Que 2:
print(df.head(6))
```

```python
#Que 3:
duplicates = df.duplicated().sum()
print("Number of duplicate records:", duplicates)
df = df.drop_duplicates()
```

```python
#Que 4:
avg_math = df['math score'].mean()
avg_reading = df['reading score'].mean()
avg_writing = df['writing score'].mean()
print("Average Scores:")
print("Math:", avg_math)
print("Reading:", avg_reading)
print("Writing:", avg_writing)
```

```python
#Que 5:
below_average = df[df['math score'] < avg_math]
print("Below average mathScore:",below_average)
```

```python
#Que 6:
count_students = df.groupby(['gender', 'parental level of education']).size()
print(count_students)
```

```python
#Que 7:
df['Total_Score'] = df['math score'] + df['reading score'] + df['writing score']
```

```python
#Que 8:
top_students = df.sort_values(by='Total_Score', ascending=False)
print(top_students.head(10))
```

```python
#Que 9:
plt.hist(df['math score'], bins=15, color='skyblue', edgecolor='black')
plt.title("Histogram of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Number of Students")
plt.grid()
```

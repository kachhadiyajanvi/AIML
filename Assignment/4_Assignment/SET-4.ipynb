```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Que 1:
data = pd.read_csv("ML Datasets/movies.csv")
print(data.info())
```

```python
#Que 2:
print(data.tail())
```

```python
#Que 3:
movies_after_2015 = data[data['Year'] > 2015]
print(movies_after_2015)
```

```python
#Que 4:
avg_rating = data['Rating'].mean()
print("Average IMDB Rating:", avg_rating)
```

```python
#Que 5:
top5_movies = data.sort_values(by='Rating', ascending=False).head(5)
print(top5_movies[['Title','Rating']])
```

```python
#Que 6:
director_count = data['Director'].value_counts()
top_director = director_count.idxmax()

print("Director with maximum movies:", top_director)
print("Number of movies:", director_count.max())
```

```python
#Que 7:
runtime_std = data['Runtime (Minutes)'].std()
print("Standard Deviation of Runtime:", runtime_std)
```

```python
#Que 8:
avg_revenue = data['Revenue (Millions)'].mean()

high_revenue_movies = data[data['Revenue (Millions)'] > avg_revenue]

print(high_revenue_movies[['Title','Revenue (Millions)']])
```

```python
#Que 9:
plt.scatter(data['Runtime (Minutes)'], data['Rating'], color='blue')

plt.title("Runtime vs Rating")
plt.xlabel("Runtime (minutes)")
plt.ylabel("IMDB Rating")

plt.show()
```

```python
#Que 10:
top5_directors = data['Director'].value_counts().head(5)

top5_directors.plot(kind='bar', color='green')

plt.title("Top 5 Directors by Movie Count")
plt.xlabel("Director")
plt.ylabel("Number of Movies")

plt.show()
```

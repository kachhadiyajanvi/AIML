```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('studentData.csv')
df
```

```python
df["Obtained marks"] = df["PoALML"] + df["ALML"] + df["Python"] + df["PoPython"]
df["Percentage"] = df["Obtained marks"] / df['Total Marks'] * 100
df
```

```python
plt.barh(df["Name"], df["Obtained marks"], color='violet', height = 0.5)
plt.title('Student Marks')
plt.xlabel('Student Name')
plt.ylabel('Obtained Marks')
plt.show()
```

```python
def show_marks(pct):
    total = sum(df['Obtained marks'])
    mark = int(round(pct * total/ 100))
    return str(mark)

plt.pie(df["Obtained marks"], labels = df["Name"], autopct=show_marks, radius = 1)
# plt.legend(title = "Name")
plt.show()
```

```python
# Set positions for the bars
bar_width = 0.2  # Width of each bar
index = np.arange(len(df))  # The positions of the bars

# Create figure and axes
plt.figure(figsize=(10, 6))

# Plot bars for each subject (stacked next to each other)
plt.barh(index, df['PoALML'], bar_width, label='PoALML', color='violet')
plt.barh(index + bar_width, df['ALML'], bar_width, label='ALML', color='lightblue')
plt.barh(index + 2 * bar_width, df['Python'], bar_width, label='Python', color='lightgreen')
plt.barh(index + 3 * bar_width, df['PoPython'], bar_width, label='PoPython', color='lightcoral')

# Adding titles and labels
plt.title('Student Marks Subject-wise')
plt.xlabel('Marks')
plt.ylabel('Student Name')
plt.yticks(index + 1.5 * bar_width, df['Name'])  # Centering the student names

# Display legend
plt.legend()
plt.show()
```

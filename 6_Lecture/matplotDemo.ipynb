```python
x = np.array(["CWS", "ALML", "POALML", "WPPython", "CN", "UI/UX"])
y = np.array([90,50,40,60,70,80])

plt.pie(y, labels = x, startangle = 90, explode = [0,0,0,0,0.1,0], shadow = True, colors = ["blue","orange","green","red","purple","brown"])
plt.show()
```

```python
x = np.array(["CWS", "ALML", "POALML", "WPPython", "CN", "UI/UX"])
y = np.array([90,50,40,60,70,80])
myExplode = [0,0,0,0,0.1,0]
plt.pie(y, labels = x, startangle = 90, explode = myExplode, shadow = True)
plt.legend(title = "Subject Name")
plt.show()
```

```python
import matplotlib.pyplot as plt
```

```python
x=[0,2,4,6,8]
y=[0,4,16,36,64]

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'o')
plt.show()
```

```python
plt.plot(x,y,'.')
plt.show()
```

```python
plt.plot(x,y,marker=".")
plt.show()
```

```python
plt.plot(y)
plt.show()
```

```python
plt.plot(x,y,'*', linestyle="-.")
plt.show()
```

```python
plt.plot(x,y,'*', linestyle=":")
plt.show()
```

```python
plt.plot(x,y,'*', linestyle="-")
plt.show()
```

```python
plt.plot(x,y,'*', linestyle="dashed")
plt.show()
```

```python
plt.plot(x,y,'*', linestyle="dashed", color="y")
plt.show()
```

```python
plt.plot(x,y,'*', linestyle="dashed",color="y", ms=12)
plt.show()
```

```python
plt.plot(x,y,'*', linestyle="dashed",color="y", ms=20, mec="r")
plt.show()
```

```python
plt.plot(x,y,'*', linestyle="dashed",color="g", ms=20, mec="b", mfc="y")
plt.show()
```

```python
plt.plot(x,y,'*', linestyle="dashed",color="g", ms=20, mec="b", mfc="y", lw=2)
plt.show()
```

```python
plt.plot(x,y,'*', linestyle="dashed",color="g", ms=20, mec="b", mfc="y", lw=2)
plt.title("Line Chart Demo")
plt.show()
```

```python
plt.plot(x,y,'*', linestyle="dashed",color="g", ms=20, mec="b", mfc="y", lw=2)
plt.title("Line Chart Demo")
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.show()
```

```python
plt.plot(x,y,'*', linestyle="dashed",color="g", ms=20, mec="b", mfc="y", lw=2)
font1={'family':'sans-serif', 'color':'b', 'size': 16,'weight': 600}
font2={'family':'serif', 'color':'r', 'size': 12}
plt.title("Line Chart Demo", fontdict = font1, loc= 'right')
plt.xlabel("X - axis", fontdict=font2)
plt.ylabel("Y - axis", fontdict=font2)
plt.show()
```

```python
x1=[0,1,2,3]
y1=[2,4,1,10]
x2=[0,1,2,3]
y2=[3,5,1,15]

plt.plot(y1)
plt.plot(y2)
plt.show()
```

```python
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```python
plt.plot(x1,y1,x2,y2)
plt.show()
```

```python
plt.plot(x1,y1,x2,y2)
plt.title("Line Chart Demo",loc= 'right')
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.grid()
```

```python
# particular grid want
plt.plot(x1,y1,x2,y2)
plt.title("Line Chart Demo",loc= 'right')
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.grid(axis='y') #parameter 'x','y' and 'both'
```

```python
# Grid color
plt.plot(x1,y1,x2,y2)
plt.grid(axis='both', color='lightblue')
```

```python
# Grid style
plt.plot(x1,y1,x2,y2)
plt.grid(axis='both', color='blue',linestyle=':')
```

```python
# grid line weight
plt.plot(x1,y1,x2,y2)
plt.grid(axis='both', color='blue',linestyle=':',lw=0.7)
```

```python
# subplot chart
# 1- Row
# 2- column
# 3- index

plt.subplot(2,1,1)
plt.plot(x1,y1)
plt.subplot(2,1,2)
plt.plot(x2,y2)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x1,y1)
plt.subplot(1,2,2)
plt.plot(x2,y2)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.grid()
plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.grid()
plt.subplot(2,2,3)
plt.plot(x2,y2)
plt.grid()
plt.subplot(2,2,4)
plt.plot(x1,y1)
plt.grid()
```

```python
plt.subplot(2,3,1)
plt.plot(x1,y1)
plt.title("subPlot 1", loc="right")
plt.grid()
plt.subplot(2,3,2)
plt.plot(x2,y2)
plt.title("subPlot 2", loc="right")
plt.grid()
plt.subplot(2,3,3)
plt.plot(x2,y2)
plt.title("subPlot 3", loc="right")
plt.grid()
plt.subplot(2,3,4)
plt.plot(x1,y1)
plt.title("subPlot 4", loc="right")
plt.grid()
plt.subplot(2,3,5)
plt.plot(x2,y2)
plt.title("subPlot 5", loc="right")
plt.grid()
plt.subplot(2,3,6)
plt.plot(x2,y2)
plt.title("subPlot 6", loc="right")
plt.suptitle("Super Title")
plt.grid()
```

```python
import matplotlib.pyplot as plt
```

```python
categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4', 'Category 5']
men_heights = [175, 178, 180, 182, 177]  # heights for men (in cm)
women_heights = [160, 162, 165, 167, 170]  #heights for women (in cm)

plt.plot(categories, men_heights, label='Men', marker='o')  # Men line
plt.plot(categories, women_heights, label='Women', marker='s')  # Women line

plt.xlabel('Categories')
plt.ylabel('Height (cm)')
plt.title('Comparison of Men and Women Heights')
plt.grid()
```
